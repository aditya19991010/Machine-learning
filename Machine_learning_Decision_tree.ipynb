### Renaming

```python
dataset.rename(columns = {'1':'initial_payment'}, inplace = True)
dataset.rename(columns = {'2':'last_payment'}, inplace = True)
dataset.rename(columns = {'3':'credit_score'}, inplace = True)
dataset.rename(columns = {'4':'house_number'}, inplace = True)
dataset.rename(columns = {'Unnamed: 5':'result'}, inplace = True)
```

```python
dataset.head()
```

## Training data

```python
#Seprating target variables

X = dataset.values[:, 0:4]
y = dataset.values[:, 5]
```

```python
#Splitting dataset into test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#Function to perform training with entropy

clf_entropy = DecisionTreeClassifier(criterion= "entropy", random_state= 100, max_depth=6, min_samples_leaf=5,)
#max_depth = shows the 3 layers of decision making

clf_entropy.fit(X_train, y_train)
```

```python
y_pred_en = clf_entropy.predict(X_test)
```

### Check accuracy

```python
accuracy_score(y_test,y_pred_en)
```

```python
clf_entropy.decision_path(X)
```
